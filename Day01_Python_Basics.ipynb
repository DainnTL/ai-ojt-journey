```python
# Calculate your GPA goal
# If you study 3.5 hours a day for 55 days
# and each hour of study improves your skills by 0.02 points
# starting from 0, what is your final skill score?

hours_of_study = 3.5
days = 55
imp_per_hour = 0.02
Final_skill_score = round(imp_per_hour * hours_of_study * days, 2)
print("Final_skill_score :",Final_skill_score)
print("As percentage:", round(Final_skill_score * 100, 2), "%" )
```
